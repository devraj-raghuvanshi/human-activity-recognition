```python
import numpy as np
import pandas as pd
import joblib
import matplotlib.pyplot as plt
```

```python
logreg_results = joblib.load('../results/logreg_results.pkl')
knn_results = joblib.load('../results/knn_results.pkl')
rfc_results = joblib.load('../results/rfc_results.pkl')
xgb_results = joblib.load('../results/xgb_results.pkl')
```

```python
df = pd.read_csv('../data/final_data.csv', index_col=0)
```

```python
df.head()
```

```python
from sklearn.metrics import f1_score, accuracy_score

def calc_baseline_performance(y_true):
    performance = {}
    classes, counts = np.unique(y_true, return_counts=True)
    majority_class = classes[np.argmax(counts)]
    y_pred = np.full_like(y_true, majority_class)
    performance['accuracy'] = accuracy_score(y_true, y_pred)
    performance['macro_f1'] = f1_score(y_true, y_pred, average="macro")

    return performance
```

```python
def calc_model_performance(model_, results):
    if model_ == 'xgb':
        acc_all_random_states = xgb_results['test_scores_acc']
        macro_f1_all_random_states = xgb_results['test_scores_macro_f1']
    else:
        acc_all_random_states = []
        macro_f1_all_random_states = []
        for i in range(len(results)):
            acc = accuracy_score(results[i]['y_test'], results[i]['y_test_pred'])
            acc_all_random_states.append(acc)
            macro_f1_all_random_states.append(results[i]['best_test_score']['score'])

    mean_test_score = {}
    std_test_score = {}

    mean_test_score['accuracy'] = np.mean(acc_all_random_states)
    std_test_score['accuracy'] = np.std(acc_all_random_states)

    mean_test_score['macro_f1'] = np.mean(macro_f1_all_random_states)
    std_test_score['macro_f1'] = np.std(macro_f1_all_random_states)
    
    # print(model_)
    # print(f"test accuracy: {mean_test_score['accuracy']:.4f} +/- {std_test_score['accuracy']:.4f}")
    # print(f"test macro f1: {mean_test_score['macro_f1']:.4f} +/- {std_test_score['macro_f1']:.4f}\n")

    return mean_test_score, std_test_score
```

### Results

#### bar plot: accuracy + macro f1

```python
def plot_model_performance(models, f1_scores, accuracy_scores, baseline_f1, baseline_accuracy):
    
    f1_mean_scores = [f1_scores[model][0] for model in models]
    f1_std_scores = [f1_scores[model][1] for model in models]
    
    accuracy_mean_scores = [accuracy_scores[model][0] for model in models]
    accuracy_std_scores = [accuracy_scores[model][1] for model in models]

    plt.figure(figsize=(11.75, 5))
    x = np.arange(len(models))

    bars_f1 = plt.bar(x - 0.2, f1_mean_scores, yerr=f1_std_scores, capsize=5, width=0.4, edgecolor='black',
                      label='Macro F1')

    bars_acc = plt.bar(x + 0.2, accuracy_mean_scores, yerr=accuracy_std_scores, capsize=5, width=0.4, edgecolor='black',
                       label='Accuracy')

    plt.axhline(y=baseline_f1, color='black', linestyle='--', label=f'Baseline Macro F1 ({baseline_f1:.2f})')
    plt.axhline(y=baseline_accuracy, color='red', linestyle='--', label=f'Baseline Accuracy ({baseline_accuracy:.2f})')
    
    plt.ylim(0, 1.1) 

    for i, bar in enumerate(bars_f1):
        mean = f1_mean_scores[i]
        std = f1_std_scores[i]
        plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.04, 
                 f"{mean:.2f}±{std:.2f}", ha='center', va='bottom', fontsize=10)

    for i, bar in enumerate(bars_acc):
        mean = accuracy_mean_scores[i]
        std = accuracy_std_scores[i]
        plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.04, 
                 f"{mean:.2f}±{std:.2f}", ha='center', va='bottom', fontsize=10)

    plt.xticks(x, models, ha='center', fontsize=12) 
    plt.yticks(fontsize=12)
    plt.xlabel('Model', fontsize=13)
    plt.ylabel('Score', fontsize=13)
    plt.title('Model Performance Comparison', fontsize=14)
    plt.legend(loc='upper left', bbox_to_anchor=(1.05, 1)) 
    plt.tight_layout()
    # plt.savefig("../figures/model_performance.png", dpi=300, bbox_inches='tight')
    plt.show()
```

```python
models = ['Logistic Regression', 'KNN', 'Random Forest', 'XGBoost']

f1_scores = {
    'Logistic Regression': (calc_model_performance('logreg', logreg_results)[0]['macro_f1'], calc_model_performance('logreg', logreg_results)[1]['macro_f1']),
    'KNN': (calc_model_performance('knn', knn_results)[0]['macro_f1'], calc_model_performance('knn', knn_results)[1]['macro_f1']),
    'Random Forest': (calc_model_performance('rfc', rfc_results)[0]['macro_f1'], calc_model_performance('rfc', rfc_results)[1]['macro_f1']),
    'XGBoost': (calc_model_performance('xgb', xgb_results)[0]['macro_f1'], calc_model_performance('xgb', xgb_results)[1]['macro_f1'])
}
accuracy_scores = {
    'Logistic Regression': (calc_model_performance('logreg', logreg_results)[0]['accuracy'], calc_model_performance('logreg', logreg_results)[1]['accuracy']),
    'KNN': (calc_model_performance('knn', knn_results)[0]['accuracy'], calc_model_performance('knn', knn_results)[1]['accuracy']),
    'Random Forest': (calc_model_performance('rfc', rfc_results)[0]['accuracy'], calc_model_performance('rfc', rfc_results)[1]['accuracy']),
    'XGBoost': (calc_model_performance('xgb', xgb_results)[0]['accuracy'], calc_model_performance('xgb', xgb_results)[1]['accuracy'])
}

y_test = rfc_results[-1]['y_test']  # using rfc here, but will be the same for all models
baseline_f1 = calc_baseline_performance(y_test)['macro_f1']
baseline_acc = calc_baseline_performance(y_test)['accuracy']

plot_model_performance(models, f1_scores, accuracy_scores, baseline_f1, baseline_acc)
```

#### Confusion Matrix

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix

def plot_confusion_matrix(y_true, y_pred, classes):
    cm = confusion_matrix(y_true, y_pred)

    plt.figure(figsize=(6, 4.5))
    sns.heatmap(cm, annot=True, fmt='g', cmap='Purples_r', xticklabels=classes, yticklabels=classes, cbar=True)
    
    plt.title('Confusion Matrix', fontsize=14)
    plt.xlabel('Predicted Label', fontsize=13)
    plt.ylabel('True Label', fontsize=13)
    plt.xticks(rotation=0, fontsize=12)
    plt.yticks(rotation=0, fontsize=12)
    plt.tight_layout()
    # plt.savefig("../figures/confusion_matrix_xgb.png", dpi=300, bbox_inches='tight')
    plt.show()
```

```python
def plot_confusion_matrix(y_true, y_pred, classes, normalize=True):
    cm = confusion_matrix(y_true, y_pred)
    
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]  # Normalize by row

    plt.figure(figsize=(6, 4.5))
    sns.heatmap(cm, annot=True, fmt='.2f' if normalize else 'g', cmap='Purples_r', 
                xticklabels=classes, yticklabels=classes, cbar=True)
    
    title = 'Normalized Confusion Matrix' if normalize else 'Confusion Matrix'
    plt.title(title, fontsize=14)
    plt.xlabel('Predicted Label', fontsize=13)
    plt.ylabel('True Label', fontsize=13)
    plt.xticks(rotation=0, fontsize=12)
    plt.yticks(rotation=0, fontsize=12)
    plt.tight_layout()
    # plt.savefig("../figures/confusion_matrix_xgb.png", dpi=300, bbox_inches='tight')
    plt.show()
```

```python
model_results = xgb_results

y_true = model_results['y_test_true'][1]
y_pred = model_results['y_test_pred'][1]
classes = np.unique(df['label'])

plot_confusion_matrix(y_true, y_pred, classes)
```

### Global Feature Importance

#### Perturbation Feature Importance

```python
random_state_idx = 1

model = xgb_results['best_models'][random_state_idx]
X_test = xgb_results['X_test_transformed'][random_state_idx]
X_test = pd.DataFrame(X_test, columns=df.columns[-37:])
y_test = xgb_results['y_test_true'][random_state_idx]
y_pred = xgb_results['y_test_pred'][random_state_idx]
ftr_names = np.array(X_test.columns)

np.random.seed(42)

nr_runs = 10
scores = np.zeros([len(ftr_names),nr_runs])

test_score = f1_score(y_test, y_pred, average='macro')
print('test macro f1 =',test_score)
print('test baseline =', calc_baseline_performance(y_test)['macro_f1'])
# loop through the features
for i in range(len(ftr_names)):
    # print('shuffling '+str(ftr_names[i]))
    f1_scores = []
    for j in range(nr_runs):
        X_test_shuffled = X_test.copy()
        X_test_shuffled[ftr_names[i]] = np.random.permutation(X_test[ftr_names[i]].values)
        f1_scores.append(f1_score(y_test, model.predict(X_test_shuffled), average='macro'))
    # print('   shuffled test score:',np.around(np.mean(f1_scores),3),'+/-',np.around(np.std(f1_scores),3))
    scores[i] = f1_scores
```

```python
sorted_indcs = np.argsort(np.mean(scores,axis=1))[::-1][-10:]
plt.rcParams.update({'font.size': 11})
plt.figure(figsize=(10,5))
plt.boxplot(scores[sorted_indcs].T,labels=ftr_names[sorted_indcs],vert=False)
plt.axvline(test_score,label='Original macro f1')
xticks = np.arange(0.75, 0.87, 0.01)  # Adjust the range as per your scores
plt.xticks(xticks, fontsize=12)
plt.yticks(fontsize=12)
plt.title("Perturbation Importances (test set)", fontsize=14)
plt.xlabel('Score with perturbed feature', fontsize=13)
plt.ylabel('Feature', fontsize=13)
plt.legend()
plt.tight_layout()
# plt.savefig("../figures/perturbation_importances.png", dpi=300, bbox_inches='tight')
plt.show()
```

#### XGBoost metrics

```python
colors = ['#011f4b', '#03396c', '#005b96', '#6497b1', '#b3cde0']  # Dark to light blue
importance_types = ['gain', 'weight', 'cover', 'total_gain', 'total_cover']

random_state_idx = 1

XGB = xgb_results['best_models'][random_state_idx]
X_test = xgb_results['X_test_transformed'][random_state_idx]
df_test = pd.DataFrame(X_test, columns=df.columns[-37:])

for i, imp_type in enumerate(importance_types):
    importances = XGB.get_booster().get_score(importance_type=imp_type)
    importances = np.array([importances.get(feature, 0) for feature in importances.keys()])
    
    top_10_indices = importances.argsort()[-10:][::-1]
    top_10_features = [df_test.columns[i] for i in top_10_indices]
    top_10_importances = importances[top_10_indices]

    # Create the plot
    plt.figure(figsize=(9, 5))
    plt.barh(np.arange(len(top_10_importances)), top_10_importances, color=colors[i])
    plt.xticks(fontsize=12)
    plt.yticks(np.arange(len(top_10_importances)), top_10_features, fontsize=12)
    plt.xlabel(f'Feature Importance', fontsize=13)
    plt.ylabel('Feature', fontsize=13)
    plt.title(f'Top 10 Most Important Features ({imp_type})', fontsize=14)
    plt.tight_layout()
    plt.gca().invert_yaxis()
    # plt.savefig(f"../figures/feature_importances_xgb_{imp_type}.png", dpi=300, bbox_inches='tight')
    plt.show()
```

#### SHAP

```python
import shap
shap.initjs()
```

```python
XGB = xgb_results['best_models'][1]
```

```python
explainer = shap.TreeExplainer(XGB)
shap_values = explainer.shap_values(df_test)
print(shap_values.shape)
```

```python
class_names = xgb_results['label_encoder'].inverse_transform(np.arange(5))

feature_shap_values = np.mean(np.abs(shap_values), axis=0)  # Shape: (n_features, n_classes)

# sorting features by their combined importance
sorted_indices = np.argsort(feature_shap_values.sum(axis=1))[::-1]
sorted_feature_shap = feature_shap_values[sorted_indices][:10]
sorted_features = np.array(ftr_names)[sorted_indices][:10]

fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.8 / len(class_names)  # Adjust bar width
activity_colors = {
    'cycling': '#332288',  # Navy Blue
    'inactive': '#117733', # Green
    'running': '#CC6677',  # Red
    'stairs': '#DDCC77',   # Yellow
    'walking': '#88CCEE',  # Light Blue
}
colors = list(activity_colors.values())

for i, class_name in enumerate(class_names):
    ax.barh(
        np.arange(len(sorted_features)) + i * bar_width,
        sorted_feature_shap[:, i],
        bar_width,
        label=class_name,
        color=colors[i % len(colors)]
    )

ax.set_yticks(np.arange(len(sorted_features)) + bar_width * (len(class_names) - 1) / 2)
ax.set_yticklabels(sorted_features)
ax.invert_yaxis()
ax.set_xlabel("mean(|SHAP value|)", fontsize=13)
ax.set_ylabel("Feature", fontsize=13)
ax.set_title("Feature Importance per Class (SHAP Summary)", fontsize=14)
ax.legend(loc="lower right", title="Classes")
plt.tight_layout()
# plt.savefig("../figures/shap_global_xgb.png", dpi=300, bbox_inches='tight')
plt.show()
```
